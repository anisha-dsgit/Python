```python
import pandas as pd 
import requests 
```

```python
Link = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"

response=requests.get(Link)
with open("IA_9.txt","wb") as file:
    file.write(response.content)
    print("data saved to file ")
    
```

```python
df=pd.read_csv("IA_9.txt")
df.head(3)
```


#### Get all the cars that have an electric range of more than 100, and were made after 2015

```python
df[ (df['Electric Range'] >100) & (df['Model Year']>2015)].sort_values(by='Electric Range').head(5)
```

#### Get all TESLA cars with the model year, and model type made in Bothell City.

```python
df[(df['Make']=='TESLA') &  (df['City']=='Bothell') ].head(3)
```

#### Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python

df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility']== 'Not eligible due to low battery range'][['Model','Make']]
```
