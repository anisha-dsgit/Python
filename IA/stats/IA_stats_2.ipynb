```python
np.random.seed(42)
```

```python
mean_height=170
stdev_height=10
num_sample=1000
#make a dataset with this informationn
data_arr=np.random.normal(loc=mean_height,scale=stdev_height,size=1000)
data_heights=np.round(data_arr,decimals=2)
data_heights
```

```python
df_height=pd.DataFrame(data={'height':data_heights})
df_height
```

```python
#a. What percentage of individuals in the dataset have heights between 160 cm and 180 cm?
#percent_indiv= people meeting criteria/total population
df_filtered=df_height[(160.0<df_height['height']) & (df_height['height']<180.0)]
people_meeting_criteria=len(df_filtered)
percent_of_people_meeting_criteria=(people_meeting_criteria/1000)*100
percent_of_people_meeting_criteria
```

b. If we randomly select 100 individuals from the dataset, what is the probability
that their average height is greater than 175 cm?

```python
#taking a random sample of height from the dataset 
#returns a
sample1= np.random.choice(a=df_height['height'] ,size=100)
sample1
```

```python
#if you are working with array use np.mean()
#if you are working with dataframe then use sample.mean()
## both the approaches will give the same result but just the base frame 

# sample_mean=sample1.mean()
# sample_stdev=sample1.std()
```

```python
#prob(ht>175)= 1-pr(ht<175)
sample_mean=np.mean(sample1)
sample_stdev=np.std(sample1,ddof=1)
print(f" sample_mean and sample stdev are {sample_mean} , {sample_stdev}") 
```

```python
z_score= (sample_mean-mean_height)/(sample_stdev/np.sqrt(100))
```

```python
prob =1- stats.norm.cdf(z_score)
prob
# Hence, the probability of avg height greater than 175 cm is 0.89. 
```

#### Assuming the dataset follows a normal distribution, what is the z-score corresponding to a height of 185 cm?

```python
# sns.displot(df_height['height']) ---> shows data is normally distributed 
z_score_of_height= (185-(mean_height))/stdev_height
z_score_of_height
```
