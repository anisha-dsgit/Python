1. Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

```python
# return max freq string and its length
string1 = " write write write write write all the number number from from from from from 1 to 100" 
str_list= string1.split()
def count_unique(string):
    freq_dict={}
    count=1
    for i in str_list:
        if i in freq_dict:
            continue
            freq_dict[i]=count
        else:
            count=str_list.count(i)
            freq_dict[i]=count
    
    print(freq_dict)
    
    max_val=max(list(freq_dict.values()))
    max_val_keys= []

    for key, value in freq_dict.items():
        if value == max_val :
            max_val_keys.append(key)
    
    maxlen=0
    for i in max_val_keys:
        if len(i)>maxlen:
            print(f"key having highest frequency in string is: {i}")
            maxlen=len(i)
    
    return  f"key with max len is: {maxlen}" 
    
        
count_unique(string=string1)
```

2. Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

```python
string_1 =input("Enter any sentence ")

def string_check(input_string):
    freq_dict={}
    for i in input_string:
        char_count=1
        if i in freq_dict:
            char_count+=1
            continue
        else:
            freq_dict[i]=char_count
    return freq_dict

string_check(string_1)
```

```python
link="https://data.nasa.gov/resource/y77d-th95.json"
pd.set_option("display.max_columns",None)
df= pd.read_json(link)
df.head(3)
```

```python
df.columns
```

```python
df=df.drop([':@computed_region_cbhk_fwbd',
       ':@computed_region_nnqa_25f4'], axis=1)

df.head()
```

```python
json_location=pd.json_normalize(df['geolocation'])
point_coordinates=json_location['coordinates']
point_coordinates
```

```python
#converted a series of just point coordinates to another df column specifically containing list of point coordinates
df['point_coordinates']=point_coordinates
```

```python
df=df.drop('geolocation',axis=1)
df.head(2)
```

```python
df['year']=df['year'].str[0:4]
df.head(3)
```

```python
#trying to make the coordinates in integer format as they are currently in float format 
df['point_coordinates'][0]=[int(x) if isinstance(x, (int, float)) else x for x in df['point_coordinates'][0]]
```

5. Write a program to download the data from the given API link and then extract the following data with
proper formatting.
Link - http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes

```python
import json
import pandas as pd
import requests
url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

response= requests.get(url=url)
json_data= response.json()
json_data
```

```python
episode_data_Df=pd.json_normalize(json_data)
episode_data_Df
```

```python
embedded_episodes=pd.json_normalize(episode_data_Df['_embedded.episodes'])
transposed_epi=embedded_episodes.transpose()
transposed_epi
```

```python
episode_df=pd.json_normalize(transposed_epi[0])
episode_df.head(3)
```
