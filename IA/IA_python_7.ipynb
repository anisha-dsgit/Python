```python
import pandas as pd
import requests
```

```python
Link = "https://data.nasa.gov/resource/y77d-th95.json"
df=pd.read_json(Link)
pd.set_option("display.max_columns",None)
df.head(3)
```

```python
df.columns
```

```python
df=df.drop([':@computed_region_cbhk_fwbd',
       ':@computed_region_nnqa_25f4'], axis=1)
df.head(3)
```

```python
df['year']=df['year'].str[0:4]
```
