#Get all Pokemons that do not have more than 2 evolutions, converting the column to list values


```python
def convert_to_list(string):
    try:
        return ast.literal_eval(string)
    except (SyntaxError, ValueError):
        return []

df['next_evolution'] = df['next_evolution'].apply(convert_to_list)
```

```python
df.head(2)
```

```python
#checking for null values in the column and then transforming them 
for i in range(len(df)):
    if pd.isnull(df['next_evolution']).iloc[i]:
        df.loc[i,'total_evolutions'] =0
    else:
        df.loc[i,'total_evolutions']=len(df['next_evolution'].iloc[i])
```

```python
pd.set_option("display.max_columns",None)
df.head(3)

df['total_evolutions']=df['total_evolutions'].astype(int)
```

```python
df['total_evolutions'].unique()
```

```python

filtered_df=df[df['total_evolutions']<=2]
pokemon_with_less_than_2evolutions=len(filtered_df)
pokemon_with_less_than_2evolutions
```

```python
import json

def convert_to_list(string):
    try:
        return json.loads(string)
    except (json.JSONDecodeError, TypeError):
        return []
    
for i in range(len(df)):
    if df['type'].iloc[i]==None:
        df.loc[i,'converted_type_col']=0
    else:
        converted_list=convert_to_list(df['type'].iloc[i])
        df.loc[i,'converted_type_col']=converted_list
        
```
