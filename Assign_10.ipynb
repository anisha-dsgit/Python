1. shutil.copy() vs shutil.copytree() , 

shutil.copy() -> copies the same file to another place. parameters source - takes the file path, destination takes the path of the destination. 

shutil.copytree()-> copies the entire folder to a different place and if you specify a name, it changes the name also.  
shutil.copytree(r"D:\ML",r"C:\Users\LENOVO\OneDrive\Desktop\mlc")
