RGBA tuples are 4-tuples where the respective tuple components represent red, green, blue, and alpha (opacity) values for a color.

Each value is a floating point number between 0.0 and 1.0.

For example, the tuple (1, 0, 0, 1) represents an opaque red, while (0, 1, 0, 0.5) represents a half transparent green.

```python
from PIL import ImageColor
ImageColor.getcolor("chocolate","RGBA")
#used to get the rgba format for one color only 
```

## Process to paste one part over another on an image 
1. if we want to exclude a part of the image from whichever corner or anywhere on the image,we take coordinates in box tuple        form. 
2. make the background black or white of the box obtained from the box tuple 
3. we then paste this box on the original image 
4. this make the part of the image go away. 
5. POINT TO REMEMBER- as images are in pixel form and each pixel is in square form that is why we use box tuple to work on the      images. 


```python
from PIL import Image
img_obj=Image.open("D:\ML\jeanpage.png")
img_obj.size

#we make a quarter object to be left out
img_obj2=Image.new('RGBA',(378,576),'White')
img_obj2.save("background.png")
img_obj2.size

img_obj.paste(img_obj2,(378,576))
img_obj.save("newphoto.png")
```
